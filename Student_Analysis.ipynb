```python
df['Average'] = df[['math percentage',
                    'reading score percentage',
                    'writing score percentage']].mean(axis=1)
```

```python
print(df.shape)
df.dtypes
```

```python
df['Rank'] = df['Average'].rank(ascending=False)
```

```python
topper=df[df['Rank']==1]
print(topper)
```

```python
print(df['Rank'].head())
```

```python
topper = df.loc[df['Average'].idxmax()]
print(topper)
```

```python
grouped = df.groupby('race/ethnicity')[[
    'math percentage',
    'reading score percentage',
    'writing score percentage'
]].mean()

print(grouped)
```

```python
top_each_group = df.loc[df.groupby('race/ethnicity')['Average'].idxmax()]
print(top_each_group)
```
